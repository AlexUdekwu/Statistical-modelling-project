# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python


api_base = 'https://api.yelp.com/v3/businesses/search?'
#set the keys
yelp_id = os.environ["AlexYelpsAPi"]
headers = {
            'Authorization': "Bearer %s" % yelp_id
        }

stations = [{
    'latitude': 49.262487,
    'longitude': -123.114397,
    # 'bikes': 5
}]
yelp_stations = {}
for station in stations:
    lat = station['latitude']
    lng = station['longitude']

    resp = requests.get(
        api_base + 'term=bike+stations&latitude=' + str(lat) + '&longitude=' + str(lng) + '&radius=1000',
        headers=headers
    )
    data = resp.json()
    yelp_stations = data['businesses']

# print(json.dumps(yelp_stations[0], indent=2))  
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
import pandas as pd

places_of_interest = []
for place in yelp_stations:
    places_of_interest.append({
        'name': place['name'],
        'rating': place['rating'],
        "latitude": place['coordinates']['latitude'],
        "longitude": place['coordinates']['longitude']
    })

# print(places_of_interest[0])

yelps_df = pd.DataFrame(places_of_interest)
yelps_df
```

```python
yelps_df['location']=[0] * (yelps_df.shape[0])
yelps_df 
```

Put your parsed results into a DataFrame
